```python
prdu_sales = fact_sales[['ProductID','UnitPrice','OrderQty','Total_Sales','Qty_Group']].merge(Dim_Product ,how = 'inner')
prdu_sales
```

```python
gby_prdu_sales = prdu_sales.groupby('ProductCategory')['Total_Sales'].sum().sort_values(ascending = False)
gby_prdu_sales
```

```python
pvt_prdu_sales = prdu_sales.pivot_table(index = 'ProductCategory' , columns = 'Qty_Group' , values = 'Total_Sales' , aggfunc = 'sum')
pvt_prdu_sales
```

```python
status_sales = Dim_status.merge(fact_sales[['UnitPrice','OrderQty','Total_Sales','Qty_Group','StatusID']] , how = 'left')
status_sales
```

```python
gby_status = status_sales.groupby('Status')['StatusID'].count().sort_values(ascending = False)
gby_status
```

```python
order_dt_sales = Dim_orderdate.merge(fact_sales[['Total_Sales','order_dateID']] , how = 'left')
order_dt_sales
```

```python
order_dt_sales.info()
```

```python
order_dt_sales['OrderDate'] = pd.to_datetime(order_dt_sales['OrderDate'])
order_dt_sales
```

```python
sales_by_year = order_dt_sales.groupby('year_')['Total_Sales'].sum()
sales_by_year
```

```python
sales_by_month = order_dt_sales.groupby('month_')['Total_Sales'].sum()
sales_by_month
```

```python
Territory_sales = Dim_Territory.merge(fact_sales[['TerritoryID','Total_Sales','TaxAmt','Qty_Group']] , how = 'left')
Territory_sales
```

```python
gby_Territory = Territory_sales.groupby('Territory')['Total_Sales'].sum().sort_values()
gby_Territory
```

```python
pvt_Territory = Territory_sales.pivot_table(index = 'Territory',columns = 'Qty_Group' ,values = 'Total_Sales' , aggfunc = 'sum')
pvt_Territory
```

```python
tax_avg_byTerritory = Territory_sales.groupby('Territory')['TaxAmt'].mean()
tax_avg_byTerritory
```
